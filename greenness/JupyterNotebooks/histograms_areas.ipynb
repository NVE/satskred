```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn.objects as so
```

```python
areas = pd.read_excel('areal_buffer.xlsx')
```

```python
from matplotlib.ticker import MultipleLocator

fig, ax = plt.subplots(figsize=(15,10))

df = areas[areas.Varslingsr == 'Romsdal']
sns.distplot(df['area_poly'],  kde=True, label='Romsdal')
df = areas[areas.Varslingsr == 'Lyngen']
sns.distplot(df['area_poly'],  kde=True, label='Lyngen')

plt.legend(prop={'size': 20})
plt.title('Distribution of polygon sizes')
plt.xlabel('Size in$\ m^2 $', size=20)
plt.ylabel('Density', size=20)


ax.spines['bottom'].set_color('dimgrey')
ax.spines['top'].set_color('dimgrey') 
ax.spines['right'].set_color('dimgrey')
ax.spines['right'].set_linewidth(1)
ax.spines['left'].set_color('dimgrey')
ax.spines['left'].set_linewidth(1)
ax.xaxis.label.set_color('black')
ax.tick_params(colors='black', which='both', labelsize=20)
```

```python
df = areas[areas.Varslingsr == 'Romsdal']
sns.distplot(df['area_poly'],  kde=False, label='Romsdal')
df = areas[areas.Varslingsr == 'Lyngen']
sns.distplot(df['area_poly'],  kde=False, label='Lyngen')

plt.legend(prop={'size': 12})
plt.title('Distribution of polygon sizes')
plt.xlabel('Size in$\ m^2 $')
plt.ylabel('Frequency')
```

```python

fig, axs = plt.subplots(figsize=(15, 10))
sns.kdeplot(areas['area_poly'], label='DET', multiple="stack")
sns.kdeplot(areas['area_20'], label='B20', multiple="stack")
sns.kdeplot(areas['area_40'], label='B40', multiple="stack")
sns.kdeplot(areas['area_60'], label='B60', multiple="stack")
sns.kdeplot(areas['area_120'], label='B120', multiple="stack")

plt.legend(prop={'size': 12})
plt.title('Distribution of polygon sizes - KDE')
plt.xlabel('Size in$\ m^2 $')
plt.ylabel('Density')
```

```python


fig, ax = plt.subplots(figsize=(15, 10))

sns.histplot(areas['area_poly'], label='DET', kde=True, log_scale=True)
sns.histplot(areas['area_20'], label='B20', kde=True, log_scale=True)
sns.histplot(areas['area_40'], label='B40', kde=True, log_scale=True)
sns.histplot(areas['area_60'], label='B60', kde=True, log_scale=True)
sns.histplot(areas['area_120'], label='B120', kde=True, log_scale=True)

plt.legend(prop={'size': 20})
plt.title('')
plt.xlabel('Area [$\ m^2 $]', size=20)
plt.ylabel('Frequency', size=20)

ax.spines['bottom'].set_color('dimgrey')
ax.spines['top'].set_color('dimgrey') 
ax.spines['right'].set_color('dimgrey')
ax.spines['right'].set_linewidth(1)
ax.spines['left'].set_color('dimgrey')
ax.spines['left'].set_linewidth(1)
ax.xaxis.label.set_color('black')
ax.tick_params(colors='black', which='both', labelsize=20)
```
