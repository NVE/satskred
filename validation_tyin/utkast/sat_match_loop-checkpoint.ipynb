```python
## EDIT KM

satman = gpd.read_file("./satskred_SHP/AvalDet_man_Merge.shp")
satman["t_0"] = pd.to_datetime(satman["Date"])
satman["t_1"] = satman["t_0"]+timedelta(days=6)
satman["days_interval"] = satman.t_1 - satman.t_0

print(satman.crs)
```

```python
wys = gpd.read_file("./wyssen_SHP/wys.shp")  # apparently the shapefile does not contain a coordinate system


print(wys.crs) # not set
# The CRS for wys is not set. The data is in lat/lon while satman is in UTM33
# therefore you get no intersections and an empty "match_dict" later

# You can set the CRS manually (or correct it in your shapefile)
# if pyproj.__version__ > 2.0
wys.crs = 'epsg:4326'
wys.to_crs('epsg:32633', inplace=True)

# if pyproj.__version__ < 2.0
# wys.crs = {'init': 'epsg:4326'}
# wys.to_crs({'init': 'epsg:32633'}, inplace=True)


print(wys.crs) # should be equal to satman now

wys.to_file(r"./wyssen_SHP/wys_33.shp")
wys["d"] = pd.to_datetime(wys["avalancher"])
```

```python
wys = gpd.read_file(".data/wyssen/wys_33/wys_33.shp")
wys["d"] = pd.to_datetime(wys["avalancher"])
```

```python
sat_gdf = satman.copy(deep=True) # makes the to dataframes independent, otherwise they are only referenced
wys_gdf = wys.copy(deep=True)
```

```python
sat_gdf = satman
wys_gdf = wys
```

```python
wys_to_sat = {} #empty dictionary
match_dict = {}

for w_row in wys_gdf.itertuples():
    sat_rows = sat_gdf[(sat_gdf["t_0"] <= w_row.d) & (w_row.d <= sat_gdf["t_1"])]
    if len(sat_rows) == 0:
        wys_to_sat[w_row.Index] = "No match"
    else:
        wys_to_sat[w_row.Index] = sat_rows.index
        for i in sat_rows.index:
            _s = sat_gdf['geometry'].iloc[i]
            _w = wys_gdf['geometry'].iloc[w_row.Index]
            if _s.intersects(_w):
                try:
                    match_dict[w_row.Index].append(i)
                except KeyError:
                    match_dict[w_row.Index] = [i]

print(match_dict) # just check that it has some content
```

```python
# Does not work for satman due to "55"

print(str(match_dict[55]))
for k in match_dict.keys():
    print(k, wys_gdf['d'].iloc[k], match_dict[k])
    ax = wys_gdf.iloc[[k]].plot()
    sat_gdf.iloc[match_dict[k]].plot(ax=ax, color='r')
    plt.gcf().savefig('match_wys{0}_sat{1}.png'.format(k, str(match_dict[k])))
```
